# EJEMPLOS DE SISTEMAS SIN SOLUCIÓN, CON UNA SOLUCIÓN Y CON INFINITAS SOLUCIONES

```python
# Línea para que de las gráficas de manera interativa
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

Se tiene el sigueinte sistema de ecuaciones

$$
y=3x+5\newline
y=-1x+3\newline
y=2x+1
$$

Se prueba si este sistema de ecuaciones tiene solución, pero primero se crea los valores para x

```python
x = np.arange(-6,6)

y_1 = 3*x + 5
y_2 = -x + 3
y_3 = 2*x + 1

# Se dibuja las ecuaciones
plt.figure()
plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)
# Se agrega los limites del gráfico
plt.xlim(-8, 8)
plt.ylim(-8, 8)
# Se dibuja los ejes
plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')
plt.show()
```

Esta gráfica muestra que no hay un punto en x,y donde se toquen las tres ecuaciones al mismo tiempo, es decir no hay punto donde se cumplan las tres ecuaciones. A este sistema de ecuaciones se conoce como un sistema sobredeterminado, es decir, se tiene mas ecuaciones que variables, por lo tanto no se tiene ninguna solución.

```python
x = np.arange(-6,6)
y_2 = -x + 3
y_3 = 2*x + 1

# Se dibuja las ecuaciones
plt.figure()
plt.plot(x, y_2)
plt.plot(x, y_3)
# Se agrega los limites del gráfico
plt.xlim(-8, 8)
plt.ylim(-8, 8)
# Se dibuja los ejes
plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')
plt.show()
```

En este caso, se muestra que las dos ecuaciones se tocan en un solo punto, por lo que el sistema de ecuaciones tiene solución en un único punto. Este es un sistema determinado, donde el número de ecuaciones es igual a las variables

```python
x = np.arange(-6,6)
y_3 = 2*x + 1

# Se dibuja las ecuaciones
plt.figure()
plt.plot(x, y_3)
# Se agrega los limites del gráfico
plt.xlim(-8, 8)
plt.ylim(-8, 8)
# Se dibuja los ejes
plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')
plt.show()
```

En esta gráfica cualquier punto es solución del problema, por la tanto cuando se tiene un número de ecuaciones menor a las variables, se tienen infinitas soluciones. Este es un sistema inderterminado

Por tanto

$$
ecuaciones> variables \rightarrow  Sobredeterminado \newline
ecuaciones= variables \rightarrow  Determinado \newline
ecuaciones< variables \rightarrow  Indeterminado \newline
$$
