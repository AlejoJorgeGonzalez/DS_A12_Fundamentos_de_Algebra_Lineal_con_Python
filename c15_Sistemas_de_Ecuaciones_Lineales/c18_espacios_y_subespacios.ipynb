# QUÉ ES UN ESPACIO Y UN SUBESPACIO

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
# se crean los vectores
v1 = np.array([1,1])
v2 = np.array([-1,-1])
```

Estos vectores pueden generar como r2 espacio

```python
# Se cambia el valor de los escalares
for a in range(-10,10):
    for b in range(-10,10):
        plt.scatter(v1[0]*a + v2[0]*b, v1[1]*a + v2[1]*b, 
                    marker='.', color='orange')

plt.xlim(-25,25)
plt.ylim(-25,25)
plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')
plt.show()
```

Con v1=(1,1) y v2=(-1,-1), por más que se cambie los valores de los escalares, siempre va a formar una linea, ya que uno de los vectores se puede expresar como el otro vector multiplicado por un escalar, es decir de un vector se puede generar el otro vector. Por tanto para este caso, al tener vectores para generar r2 se genero un sibespacio de r1, el tener o no un subespacio depende de donde nos paramos y como lo vemos.

Ahora se genera otro v1 y v2.

```python
# Se genera nuevos vectores
v1 = np.array([1,0])
v2 = np.array([2,-3])
```

```python
# Se cambia el valor de los escalares
for a in range(-10,10):
    for b in range(-10,10):
        plt.scatter(v1[0]*a + v2[0]*b, v1[1]*a + v2[1]*b, 
                    marker='.', color='orange')

plt.xlim(-25,25)
plt.ylim(-25,25)
plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')
plt.show()
```

Para este caso si se genera el espacio r2 en su totalidad, ya que un vector no se puede escribir como la combinación del otro vector. Sin embargo r2 es un subespacio con r3 y esto se puede observar a continuación

```python
# Librería necesaria para graficar en 3D
from mpl_toolkits.mplot3d import Axes3D

# Se crea los vectores en 3D
v1 = np.array([1, 0, 0]) 
v2 = np.array([2, -3, 0])

fig = plt.figure()
# La figura va a ser 3D
ax = fig.add_subplot(111, projection='3d')

for a in range(-10,10):
    for b in range(-10,10):
        ax.scatter(v1[0]*a + v2[0]*b, v1[1]*a + v2[1]*b, 
                    v1[2]*a + v2[2]*b, 
                    marker='.', color='orange')

ax.set_xlabel('eje x')
ax.set_ylabel('eje y')
ax.set_zlabel('eje z')
plt.show()
```

En esta gráfica se observa, que a pesar de tener vectores en r3, solo se puede generar un 2, por lo que estos vectores realizan un subconjunto de r3.

En álgebra líneal se le conoce como un hiperplano al subconjunto de una dimensión menos a la que se esta trabajando, por tanto r2 es un hiperplano de r3 y una línea es un hiperplano de r2
